## Numerical data scaling

```python
import timeit # To measure the performance of the models
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model #Import data set
```

```python
# Download dataset
X, y = datasets.load_diabetes(return_X_y=True)
# Dimension transformation, to fit the preprocessing format
raw = X[:,None, 2]
```

Escalation rules

```python
# max - min
max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw - max_raw -min_raw)/(max_raw - min_raw) # define scaled variables

# normalization Z-score

avg = np.average(raw)
std = np.std(raw)
z_scaled = (raw - avg)/std

fig, axs = plt.subplots(3, 1, sharex=True)

axs[0].hist(raw)
axs[1].hist(scaled)
axs[2].hist(z_scaled)
```

```python
# training models
def train_raw():# training function
    linear_model.LinearRegression().fit(raw, y) # use linear regression to scare the raw data, and

def train_scaled(): # training function
    linear_model.LinearRegression().fit(scaled, y) # linear regression training with scaled data

def train_z_scaled(): # training function
    linear_model.LinearRegression().fit(z_scaled, y) # linear regression training for Z-score scaled data
```

```python
raw_time = timeit.timeit(train_raw, number = 100) #calculate the time of the train_raw function 100 times
scaled_time = timeit.timeit(train_raw, number = 100)
z_scaled_time = timeit.timeit(train_raw, number = 100)
print('trainning time for raw data : {} '.format(raw_time))
print('trainning time for scaled data : {}'.format(scaled_time))
print('trainning time for z_scaled data : {}'.format(z_scaled_time))
```

* max-min scaling: best for evenly distributed data

* z-score scaling: best for "normally" distributed data (Gaussian bell shape)

Covariance matrix

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = sns.load_dataset('iris')
```

```python
# graph to visualize covariance
sns.pairplot(iris, hue = 'species')
```

```python
# data scaler
scaler = StandardScaler()

scaled = scaler.fit_transform(
    iris[['sepal_length','sepal_width','petal_length','petal_width']]
) # which will give an array of normalized numbers (subtract the average and divide by the standard deviation)
scaled.T
```

```python
covariance_matrix = np.cov(scaled.T) # makes the matrix with the 4x4 variables that we have written previously
covariance_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix, # data
                 cbar=True, # show color scale
                 annot=True, # show data
                 square=True, # show matriz in a grill
                 fmt='-2f',
                 annot_kws={'size':12},# font size
                 yticklabels=['sepal_length','sepal_width','petal_length','petal_width'],
                 xticklabels=['sepal_length','sepal_width','petal_length','petal_width']
                 )
```

```python
# used with standardized data
sns.jointplot(x=scaled[:,2], y=scaled[:,3])
```

Decomposition into vectors and eigenvalues: https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html

```python
# eigenvalues = valor propio
# calculate the decomposition of vectors and eigenvalues of a matrix
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
# Like this we obtain the eigenvalues, the values that stretches the most in the center
# in relation to where there is more sample data
eigen_values
```

```python
# where in relation to the covariance matrix, we have the vectors point where there is more data
eigen_vectors
```

```python
# multiply the 4 eigenvalues as a percentage to see where the greatest amount of variance is
variance_explained =[]
for i in eigen_values:
  variance_explained.append((i/sum(eigen_values))*100)

print (variance_explained)
```

```python
from sklearn.decomposition import PCA
```

```python
# Adjust the PCA function so that it eliminates two of the four dimensions/components
pca  = PCA(n_components=2)
# Apply the PCA function to adjust the dimensionality of the matrix to the scaled data
pca.fit(scaled)
```

```python
# See which addresses capture the most variance "data"
pca.explained_variance_ratio_
```

```python
# Define scaled variables
reduced_scaled = pca.transform(scaled)
```

```python
# Se aprecia que los data point tienen loso dos componentes ya que tienen solo 2 datos
# It can be seen that the data points have both components since they only have 2 dimensions.
reduced_scaled
```

```python
iris['pca_1'] = reduced_scaled[:, 0]
iris['pca_2'] = reduced_scaled[:, 1]
sns.jointplot(x = iris['pca_1'], y= iris['pca_2'], hue=iris['species'])
```
